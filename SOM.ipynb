```python
clean_data = pd.read_csv('http://10.237.200.3/loyalty_sub_info2.dsv', encoding = 'utf8')
```

```python
clean_data1 = clean_data.loc[:,clean_data.isnull().mean()<=0]
```

```python
remove = []

for col in clean_data1.columns:
    try:
        _ = clean_data1[col].astype(float)
    except ValueError:
        remove.append(col)
        pass
clean_data2 = clean_data1[[col for col in clean_data1.columns if col not in remove]]
```
